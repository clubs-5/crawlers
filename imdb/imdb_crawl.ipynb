```python
from urllib import request
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/feature/genre/?ref_=nv_tv_gr'  
#拜訪網頁
res = request.urlopen(url) 
#轉換soup
soup = BeautifulSoup(res, 'html.parser') 
#爬取大區塊
soup1 =soup.select('.ab_ninja') 
#從圖片爬取網址
for t in soup1:
    imageurl = t.select('.image')
    img_url = list(map(lambda x: x.select_one('a'), imageurl))
#進入網址
    for i in img_url:
        res_img = request.urlopen(i.get('href'))
        soup_url = BeautifulSoup(res_img, 'html.parser')
      #  for title in soup_url.select('h3[class="lister-item-header"]'):
      #      print(title.a.text)
      #      print(title.p)
#爬取片名、片長、類別、評分、演員        
        for i in soup_url.select('div[class="lister-item-content"]'):
            try: 
                print(i.a.text) 
                print(select(i.p.select('span[class="runtime"]'))) 
                print(select(i.p.select('span[class="genre"]'))) 
                print(select(i.div.select('div[name="ir"]')))
                print(i.p.a.text)
            except AttributeError as e:
                print('Error: ', type(e), str(e))
                
            
```

```python
def select(x):
    result = ''
    for i in x:
        result += i.text
    return result
```
