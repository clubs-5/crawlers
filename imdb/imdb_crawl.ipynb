```python
from urllib import request
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/feature/genre/?ref_=nv_tv_gr'  
res = request.urlopen(url) #拜訪網頁
soup = BeautifulSoup(res, 'html.parser') #轉換soup
soup1 =soup.select('.ab_ninja') #爬取大區塊
for t in soup1: #從圖片爬取網址
    imageurl = t.select('.image')
    img_url = list(map(lambda x: x.select_one('a'), imageurl))
    for i in img_url: #進入網址
        res_img = request.urlopen(i.get('href'))
        soup_url = BeautifulSoup(res_img, 'html.parser')
      #  for title in soup_url.select('h3[class="lister-item-header"]'):
      #      print(title.a.text)
      #      print(title.p)
        for i in soup_url.select('div[class="lister-item-content"]'):
            
            print(i.a.text) #爬取片名
            print(select(i.p.select('span[class="runtime"]'))) #爬取片長
            print(select(i.p.select('span[class="genre"]'))) #爬取類別
            print(select(i.div.select('div[name="ir"]'))) #爬取評分
            
```

```python
def select(x):
    result = ''
    for i in x:
        result += i.text
    return result
```
